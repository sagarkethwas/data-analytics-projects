```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming your DataFrame is named df
categorical_features = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport',
    'StreamingTV', 'StreamingMovies', 'Contract', 'PaperlessBilling'
]

# Set plot style
sns.set(style="whitegrid")

# Create countplots for each feature
for feature in categorical_features:
    plt.figure(figsize=(6, 4))
    sns.countplot(data=df, x=feature, palette='Set2')
    plt.title(f'Countplot of {feature}')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

# summary

#This exploratory data analysis was conducted on a telecom customer dataset containing 7,043 records and 21 attributes, with the goal of understanding factors driving customer churn.

Churn Rate: Around 26–27% of customers were found to have churned, indicating a significant retention challenge.

Demographics: The gender distribution is roughly equal (~50.5% Male, ~49.5% Female), suggesting churn is not gender-specific. Senior Citizens account for about 16% of the customer base but show a higher churn rate (~41%), highlighting them as a vulnerable group.

Contract Type: Customers on month-to-month contracts make up ~55% of the base and contribute over 85% of churners, while those with one-year or two-year contracts have much lower churn rates (~11% and ~3%, respectively).

Internet Service: About 74% of customers have Internet service, with Fiber Optic users showing a much higher churn rate (~42%) compared to DSL (~18%).

Online Services: Lack of Online Security, Device Protection, and Tech Support strongly correlates with churn – customers without these add-ons are 20–25% more likely to leave.

Charges: Churn is higher among customers with monthly charges above $70, showing price sensitivity in the higher spending segment.

Tenure: Customers with tenure under 12 months contribute ~50% of churn, emphasizing the importance of improving onboarding and early engagement.

Overall, churn is primarily driven by contract type, tenure, and internet service type. The findings suggest that encouraging longer-term contracts, offering security and support add-ons, and targeted retention efforts for new and fiber optic customers could significantly reduce churn.
